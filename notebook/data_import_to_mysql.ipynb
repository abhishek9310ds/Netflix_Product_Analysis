```python
import pandas as pd
from sqlalchemy import create_engine

# MYSQL CONNECTION

username = "root"
password = "ryuk727k"
host = "localhost"
port = "3306"
database = "spotify_product_analytics"

# CREATE ENGINE

engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
)

# LOAD DATASETS

events_df = pd.read_csv("../analytics_data/events.csv")

sessions_df = pd.read_csv("../analytics_data/sessions.csv")

tracks_df = pd.read_csv("../cleaned_data/tracks_cleaned.csv")

# EXPORT TO MYSQL

events_df.to_sql(
    name='events',
    con=engine,
    if_exists='replace',
    index=False
)

sessions_df.to_sql(
    name='sessions',
    con=engine,
    if_exists='replace',
    index=False
)

tracks_df.to_sql(
    name='tracks',
    con=engine,
    if_exists='replace',
    index=False
)

print("All tables imported successfully!")
```
